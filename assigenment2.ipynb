```python
#Basics of Probability
# Write a Python program to simulate the following scenarios:  
#   a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.  
#   b. Rolling two dice and computing the probability of getting a sum of 7.  
#   Steps  
#       a. Use Python's random module for simulations.  
#       b. Implement loops for repeated trials.  
#       c. Track outcomes and compute probabilities.  


import random

def simulate_coin_tosses(num_tosses):
    heads_count = 0
    tails_count = 0
    
    for _ in range(num_tosses):
        toss = random.choice(['H', 'T'])  # Randomly choose the 'H' or 'T'
        if toss == 'H':
            heads_count += 1
        else:
            tails_count += 1
            
    probability_heads = heads_count / num_tosses
    probability_tails = tails_count / num_tosses
    
    return probability_heads, probability_tails

def simulate_dice_rolls(num_trials):
    sum_seven_count = 0
    
    for _ in range(num_trials):
        die1 = random.randint(1, 6)  # Roll the first die
        die2 = random.randint(1, 6)  # Roll the second die
        if die1 + die2 == 7:
            sum_seven_count += 1
            
    probability_sum_seven = sum_seven_count / num_trials
    
    return probability_sum_seven

# Main simulation
if __name__ == "__main__":
    # Simulate coin tosses
    num_tosses = 10000
    prob_heads, prob_tails = simulate_coin_tosses(num_tosses)
    print(f" Probability of Heads: {prob_heads:.4f}")
    print(f" Probability of Tails: {prob_tails:.4f}")

    # Simulate dice rolls
    num_dice_trials = 10000
    prob_sum_seven = simulate_dice_rolls(num_dice_trials)
    print(f" Probability of getting a sum of 7: {prob_sum_seven:.4f}")
```

```python
# 2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.  
#   Steps 
#       a. Simulate rolling a die 10 times using a loop.  
#       b. Track trials where at least one "6" occurs.  
#       c. Calculate the proportion of successful trials.

import random

def estimate_probability_of_six(num_trials, rolls_per_trial):
    success_count = 0  # Count of the trials with at least one "6"
    
    for _ in range(num_trials):
        found_six = False  # Flag to track if a "6" is found in this trial
        for _ in range(rolls_per_trial):
            if random.randint(1, 6) == 6:  # Roll the die
                found_six = True  # Set the flag if "6" is found
                break
        if found_six:
            success_count += 1  # Increment success count if "6" was found
            
    probability_at_least_one_six = success_count / num_trials  # Calculate the probability
    
    return probability_at_least_one_six

# Main block to run the function
if __name__ == "__main__":
    num_trials = 10000  # Number of trials to simulate
    rolls_per_trial = 10  # Number of rolls in each trial
    probability_six = estimate_probability_of_six(num_trials, rolls_per_trial)
    print(f"Probability of getting at least one '6' in {rolls_per_trial} rolls: {probability_six:.4f}")
```

```python
# Conditional Probability and Bayes' Theorem


# 3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag. If this process is repeated 1000 times, write a Python program to estimate:  
#   a. The probability of drawing a red ball given that the previous ball was blue.  
#   b. Verify Bayes' theorem with the simulation results.  


# Steps  
#     a. Use random sampling to simulate the process.  
#     b. Compute conditional probabilities directly from the data.  

import random

def simulate_ball_draws(num_draws):
    # Define the balls in the bag
    balls = ['red'] * 5 + ['green'] * 7 + ['blue'] * 8
    draws = []  # List to store the results of each draw

    for _ in range(num_draws):
        draw = random.choice(balls)  # Randomly draw a ball
        draws.append(draw)  # Store the result

    return draws

def compute_conditional_probability(draws):
    # Count occurrences
    count_red_given_blue = 0
    count_blue = 0

    for i in range(1, len(draws)):
        if draws[i-1] == 'blue':  # Check if the previous ball was blue
            count_blue += 1
            if draws[i] == 'red':  # Check if the current ball is red
                count_red_given_blue += 1

    # Calculate conditional probability P(Red | Blue)
    if count_blue > 0:
        probability_red_given_blue = count_red_given_blue / count_blue
    else:
        probability_red_given_blue = 0  # Avoid division by zero

    return probability_red_given_blue, count_red_given_blue, count_blue

def verify_bayes_theorem(draws):
    # Count occurrences
    count_red = draws.count('red')
    count_blue = draws.count('blue')

    # Total draws
    total_draws = len(draws)

    # Calculate P(Red), P(Blue), and P(Blue | Red)
    probability_red = count_red / total_draws
    probability_blue = count_blue / total_draws

    # Calculate P(Blue | Red) using the counts
    count_red_given_blue = 0
    for i in range(1, len(draws)):
        if draws[i] == 'red' and draws[i-1] == 'blue':
            count_red_given_blue += 1

    if count_red > 0:
        probability_blue_given_red = count_red_given_blue / count_red
    else:
        probability_blue_given_red = 0  # to Avoid the division by zero

    # Verify Bayes' theorem: P(Red | Blue) = P(Blue | Red) * P(Red) / P(Blue)
    if probability_blue > 0:
        bayes_verification = (probability_blue_given_red * probability_red) / probability_blue
    else:
        bayes_verification = 0  # to Avoid the  division by zero

    return bayes_verification

# Main the block to run the simulation
if __name__ == "__main__":
    num_draws = 1000
    draws = simulate_ball_draws(num_draws)

    # Compute conditional probability P(Red | Blue)
    probability_red_given_blue, count_red_given_blue, count_blue = compute_conditional_probability(draws)
    print(f"Estimated P(Red | Blue): {probability_red_given_blue:.4f} (Count: {count_red_given_blue} out of {count_blue})")

    # Verify the Bayes' theorem
    bayes_verification = verify_bayes_theorem(draws)
    print(f"the Bayes' theorem verification result is : {bayes_verification:.4f}")

```

```python
# Random Variables and Discrete Probability


# 4. Generate a sample of size 1000 from a discrete random variable with the following distribution:  
#   - P(X=1) = 0.25  
#   - P(X=2) = 0.35  
#   - P(X=3) = 0.4  
#   Compute the empirical mean, variance, and standard deviation of the sample.  
#   Steps  
#       a. Use numpy.random.choice() to generate the sample.  
#       b. Use numpy methods to calculate mean, variance, and standard deviation.

import numpy as np

# Define the probabilities and corresponding values
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

# Generate a sample of size 1000
sample_size = 1000
sample = np.random.choice(values, size=sample_size, p=probabilities)

# Calculate the empirical mean, variance, and standard deviation
empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)
empirical_std_dev = np.std(sample)

# Print the results
print(f"the Empirical Mean: {empirical_mean:.4f}")
print(f"the Empirical Variance: {empirical_variance:.4f}")
print(f"the Empirical Standard Deviation: {empirical_std_dev:.4f}")
```

```python
# Continuous Random Variables


# 5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
#   a. A histogram.  
#   b. A probability density function (PDF) overlay.  
#   Steps  
#       a. Use numpy.random.exponential().  
#       b. Use matplotlib to create visualizations.  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the mean and scale for the exponential distribution
mean = 5
scale = mean  # In an exponential distribution, scale is equal to the mean

# Generate 2000 random samples from the exponential distribution
samples = np.random.exponential(scale, 2000)

# Create a histogram of the samples
plt.figure(figsize=(12, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Create a range of x values for the PDF
x = np.linspace(0, max(samples), 1000)

# Calculate the PDF of the exponential distribution
pdf = (1 / scale) * np.exp(-x / scale)

# Overlay the PDF on the histogram
plt.plot(x, pdf, color='orange', lw=2, label='PDF (Exponential Distribution)')
plt.title('Histogram of Exponential Distribution Samples with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
# Central Limit Theorem


# 6. Simulate the Central Limit Theorem by following these steps  
#   a. Generate 10,000 random numbers from a uniform distribution.  
#   b. Draw 1000 samples of size n = 30.  
#   c. Calculate and visualize the distribution of sample means.  
#   Steps  
#       a. Use numpy.random.uniform().  
#       b. Plot both the uniform distribution and the sample mean distribution for comparison.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate  the 10,000 random numbers from a uniform distribution
num_random_numbers = 10000
uniform_data = np.random.uniform(low=0, high=1, size=num_random_numbers)

# Step 2: Draw the 1000 samples of size n = 30
sample_size = 30
num_samples = 1000
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(uniform_data, size=sample_size)
    sample_means.append(np.mean(sample))

# Convert sample means to a numpy array for the easier manipulation
sample_means = np.array(sample_means)

# Step 3: Visualize the distributions
plt.figure(figsize=(12, 6))

# Plot the uniform distribution
plt.subplot(1, 2, 1)
plt.hist(uniform_data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
plt.title('Distribution of Sample Means (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
